```python
import numpy as np
import pandas as pd
```

```python
column_names = ['user_id', 'item_id', 'rating', 'timestamp']
df = pd.read_csv('/content/u.data' , sep='\t',names=column_names)
```

```python
df.head()
```

```python
movie_titles = pd.read_csv('/content/Movie_Id_Titles.txt')
movie_titles.head()
```

```python
# Convert 'item_id' in df to numeric, handling potential errors
df['item_id'] = pd.to_numeric(df['item_id'], errors='coerce')

# Now perform the merge
df = pd.merge(df, movie_titles, on='item_id')
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')
%matplotlib inline
```

```python
df.groupby('title')['rating'].mean().sort_values(ascending=False).head()
```

```python
df.groupby('title')['rating'].count().sort_values(ascending=False).head()
```

```python
ratings = pd.DataFrame(df.groupby('title')['rating'].mean())
ratings.head()
```

```python
# Calculate the mean rating and number of ratings per movie
ratings = df.groupby('title')['rating'].agg(['mean', 'count'])

# Rename the 'count' column for clarity
ratings = ratings.rename(columns={'count': 'num of ratings'})

# Plot the histogram
plt.figure(figsize=(10,4))
ratings['num of ratings'].hist(bins=70)
plt.xlabel('Number of Ratings')
plt.ylabel('Frequency')
plt.title('Distribution of Number of Ratings per Movie')
plt.show()
```

```python
# Calculate the mean rating and number of ratings per movie
ratings = df.groupby('title')['rating'].agg(['mean', 'count'])

# Rename the 'count' column for clarity
ratings = ratings.rename(columns={'count': 'num of ratings'})

# Plot the histogram
plt.figure(figsize=(10,4))
# Access the 'mean' column using the new column name
ratings['mean'].hist(bins=70)
plt.xlabel('Mean Rating')
plt.ylabel('Frequency')
plt.title('Distribution of Mean Ratings per Movie')
plt.show()
```

```python
# Calculate the mean rating and number of ratings per movie
ratings = df.groupby('title')['rating'].agg(['mean', 'count'])

# Rename the 'count' column for clarity
ratings = ratings.rename(columns={'count': 'num of ratings'})

# Plot the jointplot
sns.jointplot(x='mean', y='num of ratings', data=ratings, alpha=0.5)  # Use 'mean' for x-axis
plt.show()
```

```python
moviemat = df.pivot_table(index='user_id',columns='title',values='rating')
moviemat.head()
```

```python
ratings.sort_values('num of ratings',ascending=False).head(10)
```

```python
ratings.head()
```

```python
starwars_user_ratings = moviemat['Star Wars (1977)']
liarliar_user_ratings = moviemat['Liar Liar (1997)']
starwars_user_ratings.head()
```

```python
similar_to_starwars = moviemat.corrwith(starwars_user_ratings)
similar_to_liarliar = moviemat.corrwith(liarliar_user_ratings)
```

```python
corr_starwars = pd.DataFrame(similar_to_starwars,columns=['Correlation'])
corr_starwars.dropna(inplace=True)
corr_starwars.head()
```

```python
corr_starwars.sort_values('Correlation',ascending=False).head(10)
```

```python
corr_starwars = corr_starwars.join(ratings['num of ratings'])
corr_starwars.head()
```

```python
corr_starwars[corr_starwars['num of ratings']>100].sort_values('Correlation',ascending=False).head()
```

```python
corr_liarliar = pd.DataFrame(similar_to_liarliar,columns=['Correlation'])
corr_liarliar.dropna(inplace=True)
corr_liarliar = corr_liarliar.join(ratings['num of ratings'])
corr_liarliar[corr_liarliar['num of ratings']>100].sort_values('Correlation',ascending=False).head()
```
